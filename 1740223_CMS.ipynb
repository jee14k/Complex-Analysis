```python
from sympy import *
def mtm_method_plus():
    u,v,x,y,z,c= symbols('u v x y z c')
    expr=(sin(x)+cos(x))*(exp(y))
    print("Given Expression: ",expr)
    diff_x=diff(expr,x)
    diff_y=diff(expr,y)
    print("\nDerivative wrt x: ",diff_x)
    print("\nDerivative wrt y: ",diff_y)
    diff_u=(diff_x+diff_y)/2
    diff_v=(diff_x-diff_y)/2
    print("\nDerivative of u wrt x: ",diff_u)
    print("\nDerivative of u v wrt x: ",diff_v)
    diff_z=diff_u+1j*diff_v
    diff_z=diff_z.subs(x,z)
    diff_z=diff_z.subs(y,0)
    print("\nHarmonic Conjugate f(z)=",integrate(diff_z)+c)
```
